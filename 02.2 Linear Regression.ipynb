# Linear Regression

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
plt.rcParams["figure.figsize"] = (8,4)
fig = plt.figure()
ax1=fig.subplots(1, 1)
ax1.set_xlabel(r'$\x')
ax1.set_ylabel(r'$y=wx+b$')
ax1.set_title('Data Points from Linear Equation')

x=np.arange(-1,1,0.01)
yt=3*x+2 + np.random.randn(len(x))
y=3*x+2

ax1.plot(x,yt,'bo',label=r'$yt$')
ax1.plot(x,y,'g-',label=r'$y=3x+2$')
ax1.legend()
```

We have a set of points (blue dots) that generated from model $y=3x+2$.
We want computer to learn from $y_t$ and create accurate model called $yp$.
we can model $yp=wx+b$ but we need to adjust $\theta=[w,b]$. Let
\begin{eqnarray*}
yp&=&wx+b\\
L&=&\frac{1}{2}\sum_i (yp_i-yt_i)^2\\
\theta_{t+1} &\gets& \theta_{t}-\alpha\frac{\partial L}{\partial \theta}
\end{eqnarray*}

Now we need to GD on the bias $b$, which $b_{t+1}=b_{t}-\alpha\frac{\partial L}{\partial b}$.

Let focus on $\frac{\partial L}{\partial b}$
\begin{eqnarray*}
\frac{\partial L}{\partial b}&=&\frac{1}{2n} \sum_i^n \frac{\partial L}{\partial yp}.\frac{\partial yp}{\partial b} \\
&=&\frac{1}{2n} \sum_i^n 2(yp_i-yt_i).1\\
&=&\frac{1}{n}\sum_i^n yt_i-yt_i
\end{eqnarray*}

Now we need to GD on the weight $w$, which $w_{t+1}=w_{t}-\alpha\frac{\partial L}{\partial w}$

Let focus on $\frac{\partial L}{\partial w}$
\begin{eqnarray*}
\frac{\partial L}{\partial b}&=&\frac{1}{2n} \sum_i^n \frac{\partial L}{\partial yp}.\frac{\partial yp}{\partial w} \\
&=&\frac{1}{2n} \sum_i^n 2(yp_i-yt_i).x_i\\
&=&\frac{1}{n}\sum_i^n (yt_i-yt_i).x_i
\end{eqnarray*}

```python
theta1 = 0.0
theta2 = 0.0
alpha =0.1

def yp(x):
    global theta1,theta2
    return theta1*x+theta2

plt.plot(x,yt,'bo')
for i in range(100):
    theta1 = theta1 - alpha*np.average( (yp(x)-yt)*x ) #weight
    theta2 = theta2 - alpha*np.average( (yp(x)-yt) ) #bias
    if i%10==0:
        print(theta1,theta2)
        plt.plot(x,theta1*x+theta2,'r')
```

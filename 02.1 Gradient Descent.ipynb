# Gradient Descent

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
#plt.style.use('ggplot')
```

```python
plt.rcParams["figure.figsize"] = (8,4)
fig = plt.figure()
ax1 =fig.subplots(1, 1)
ax1.set_xlabel(r'$\theta$')
ax1.set_ylabel(r'$L(\theta)$')
ax1.set_title('Loss function')

theta=np.linspace(start=-1,stop=2,num=50)
L=theta*theta-theta+1
ax1.plot(theta,L,'b',label=r'$L=\theta^2-\theta+1$')
#plt.grid()
ax1.legend()
```

## Gradient Descent(GD)
The GD algorithm is numerical method to find minmum point of the quadratic function.

\begin{equation*}
\theta_{t+1} \gets \theta_{t}-\alpha\frac{\partial L}{\partial \theta}
\end{equation*}


Here we have $L$ is a loss function
$\alpha$ is a learning rate
$\theta$ is parameter we need to adjust


```python
theta = 1.5
alpha =0.3
for i in range(10):
    theta = theta - alpha*(2*theta-1)
    print(theta)
```

```python
def loss(theta):
    return theta*theta-theta+1

plt.rcParams["figure.figsize"] = (16,4)
fig = plt.figure()
ax1, ax2, ax3=fig.subplots(1, 3)
d=[(ax1,0.1),(ax2,0.7),(ax3,1.02)]

for ax,alpha in d:
    ax.set_xlabel(r'$\theta$')
    ax.set_ylabel(r'$L(\theta)$')
    ax.set_title(r'Loss function and GD: $\alpha=%0.2f$'%alpha)
    thetai=np.linspace(start=-1,stop=2,num=50)
    ax.plot(thetai,loss(thetai),'b',label=r'$L=\theta^2-\theta+1$')
    thetas=[]
    theta =1.5
    for i in range(10):
        thetas.append(theta)
        theta = theta - alpha*(2*theta-1)
    thetas=np.array(thetas)
    ax.plot(thetas,loss(thetas),'ro--',label=r'$GD$')
    ax.legend()
```

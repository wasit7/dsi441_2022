```python
%matplotlib notebook
from mpl_toolkits import mplot3d
import numpy as np
from matplotlib import pyplot as plt
plt.style.use('ggplot')
```

```python
n=200
def normal(n):
    return np.random.randn(n)

x1=np.concatenate((normal(n)-2,normal(n)+1),axis=None)
x2=np.concatenate((normal(n)-2,normal(n)+2),axis=None)
yt=np.ones(2*n)
yt[:n]=-1
# plt.plot(x1[yt==-1],x2[yt==-1],'or')
# plt.plot(x1[yt==1],x2[yt==1],'ob')


plt.rcParams["figure.figsize"] = (8,6)
fig = plt.figure()
# ax = plt.axes(projection='3d')
ax=fig.subplots(1, 1)
ax.set_xlabel('$x1$')
ax.set_ylabel('$x2$')

ax.scatter(x=x1[yt==-1], y=x2[yt==-1], c='r', label="yt")
ax.scatter(x=x1[yt==1], y=x2[yt==1], c='b', label="yt")
# ax.legend()
```

```python
from mpl_toolkits import mplot3d
plt.rcParams["figure.figsize"] = (8,6)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('$x1$')
ax.set_ylabel('$x2$')
ax.set_zlabel('$y$');


ax.scatter3D(xs=x1[yt==-1], ys=x2[yt==-1], c='r', zs=yt[yt==-1], label="yt")
ax.scatter3D(xs=x1[yt==1], ys=x2[yt==1], c='b', zs=yt[yt==1], label="yt")
#ax.scatter3D(xs=mx1, ys=mx2, zs=g(mx1,mx2), label="g()=x2")
#ax.scatter3D(xs=mx1, ys=mx2, zs=h(mx1,mx2), label="h()=x1+x2")
ax.legend()
```
